### 2. Explore & Clean Data

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
df = pd.read_csv("student-mat.csv", sep=";")

print(f"Dataset Shape: {df.shape[0]} rows, {df.shape[1]} columns\n")
print("Missing Values per Column (Top 5):")
print(df.isnull().sum().head())

duplicates_count = df.duplicated().sum()
print(f"\nNumber of duplicate rows found: {duplicates_count}")
if duplicates_count > 0:
    df = df.drop_duplicates()
    print("Duplicates removed successfully.")

print("--- ANALYSIS RESULTS ---")
avg_g3 = df['G3'].mean()
print(f"1. Average final grade (G3): {avg_g3:.2f}")

students_above_15 = df[df['G3'] > 15].shape[0]
print(f"2. Number of students who scored above 15 in G3: {students_above_15}")

correlation = df['studytime'].corr(df['G3'])
print(f"3. Correlation between study time and final grade (G3): {correlation:.4f}")

gender_perf = df.groupby('sex')['G3'].mean()
print("\n4. Average final grade by gender:")
print(gender_perf.to_string())

plt.figure(figsize=(8,4))
sns.histplot(df['G3'], bins=10, color='skyblue', edgecolor='black', kde=True)
plt.title('Distribution of Final Grades (G3)')
plt.show()

plt.figure(figsize=(8,4))
sns.stripplot(data=df, x='studytime', y='G3', alpha=0.5, color='purple', jitter=0.2)
plt.title('Scatterplot: Study Time vs. Final Grades')
plt.show()

plt.figure(figsize=(5,4))
sns.barplot(data=df, x='sex', y='G3', errorbar=None, palette='muted')
plt.title('Average Final Grade by Gender')
plt.show()
```
